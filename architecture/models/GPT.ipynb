## 1. Overview

<!-- What is this concept? 2–4 sentences that a newcomer could understand.
     Include: what problem it solves, why it exists, where it fits in the AI landscape. -->

### **What is GPT?**
GPT models are a family of large language models based on a transformer deep learning architecture, developed by OpenAI. These foundation models power ChatGPT and other generative AI applications capable of simulating human-created output.
### **Where did it come from?**
The transformer architecture, the core technology of a GPT, was developed by researchers at Google and introduced in the paper "Attention Is All You Need," released on June 12, 2017. OpenAI was the first to apply generative pre-training to this architecture, introducing the GPT-1 model in 2018.
### **Difference with BERT and BART?**
* **GPT** is a **decoder-only** model that reads text in one direction (left to right) and predicts what comes next. It excels at generating fluent, human-like text, making it the go-to choice for chatbots, creative writing, and code generation.

* **BERT** is an **encoder-only** model that reads text in both directions simultaneously, giving it a deep understanding of context and meaning. It is ideal for tasks like sentiment analysis, question answering, and search ranking.

* **BART** is an **encoder-decoder** model that combines the best of both worlds. It first reads and understands the full input like BERT, then generates a new output sequence like GPT. This makes it particularly powerful for tasks that require both comprehension and generation, such as summarisation and machine translation.

## 2. How It Works

### 2.1 Pre-training 
GPT models are pre-trained on large datasets of unlabeled content. This is a form of self-supervised learning wherein a model is first trained on a large, unlabeled dataset to learn to generate data points.

### 2.2 Tokenization 
The raw text you type is broken into small chunks called tokens (words or parts of words), which are converted into numbers the model can process.

### 2.3 Embeddings
Those numbers are transformed into rich vectors that capture the meaning of each word in relation to others.

### 2.4 Transformer layers (Self-Attention)
GPT models consist of multiple transformer blocks. Each block applies self-attention to model long-range dependencies in text. The models are trained on massive text corpora to predict the next token autoregressively, given the previous context. Self-attention mechanisms allow GPT to weigh the importance of different parts of the input sequence when generating output text, capturing context and dependencies between words and sentences.

### 2.5 Output
A probability score is assigned to every possible next word, and the most likely one is selected. Then, the whole process repeats (the autoregressive loop shown on the left of the diagram).

## 3. Advantages & Limitations

| | Aspect | Commentary |
|--|--------|------------|
| 🟢 | **Versatility** | GPT models are general-purpose and can perform a broad range of tasks, from creating original content to writing code, summarizing text, and extracting data. |
| 🟢 | **Few-shot learning** | Few-shot learning enables fast iteration on new use cases, allowing companies to implement NLP systems faster. |
| 🟢 | **Human-like fluency** | GPT represents a notable breakthrough in natural language processing, propelling development of machines that can understand and communicate in a manner that closely resembles humans. |
| 🟢 | **Reduced engineering cost** | Pre-trained capacities reduce the data and engineering investments required to operationalize AI solutions, allowing enterprises to accelerate valuable applications. |
| 🔴 | **No true understanding** | GPT predicts tokens statistically but it does not "understand" meaning the way humans do, and can produce plausible-sounding but factually wrong answers (hallucinations). |
| 🔴 | **Computational cost** | The standard self-attention mechanism has a quadratic complexity with respect to input sequence length, making large models extremely expensive to train and run.|
| 🔴 | **Training data bias** | Because GPT learns from human-written text, it can reproduce biases, stereotypes, and misinformation present in that data. |
| 🔴 | **No real-time knowledge** | A GPT model's knowledge is frozen at its training cutoff date.|

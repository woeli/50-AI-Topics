## What is BART?
BART (Bidirectional and Auto-Regressive Transformers) is a sequence-to-sequence foundation model designed for natural language processing, developed by Meta AI (formerly Facebook AI). It is a hybrid model engineered to excel at both understanding language and generating new text.

## How does it learn?
BART is trained using a "denoising" process (high-tech version of text repair). The model is fed corrupted text—where words are deleted, sentences are shuffled, or tokens are masked. The model must then learn how to reconstruct the original, clean document.

1) The encoder analyzes the messy, corrupted input to understand the context.

2) The decoder then takes that analysis and generates the corrected text from scratch.

## Difference with GPT and BERT?
**BERT** is an encoder-only model that reads text in both directions simultaneously to deeply understand context. It is ideal for analytical tasks like search ranking and sentiment analysis, but it cannot generate new text.

**GPT** is a decoder-only model that reads text from left to right to predict the next word. It excels at generating highly fluent, human-like text, making it the go-to for chatbots and creative writing, but it lacks a separate encoding phase for deep comprehension.

**BART** is an encoder-decoder model that combines the strengths of both. It uses a BERT-like encoder to read and digest the full input, and a GPT-like decoder to generate the output. This dual architecture makes BART uniquely powerful for complex "translation" tasks that require both deep reading and writing, such as text summarization, abstracting, and machine translation.

## 2. How It Works

### Step 1: Corrupt the input
**BART** is trained by **corrupting text with an arbitrary noising function**, and learning a model to reconstruct the original text.

The noising schemes used are:
* **Token masking**: Replacing random tokens with [MASK] ← learns how to predict the token based on the rest of the sequence
* **Token deletion**: Removing random tokens ← learns to predict both the content and position
* **Text infilling**:  Removes a fixed number of contiguous tokens and replaces them with a single [MASK] token ← learn the content of the missing tokens and the number of tokens. 
* **Sentence permutation**: Randomly shuffles sentences ← learns global document structure and the logical ordering of ideas
* **Document rotation**: Rotates the document so that it begins at a randomly chosen token ← learns to identify the true beginning of a document

### Step 2: Encode bidirectionally
The **encoder** processes the entire input text **bidirectionally**, allowing it to understand the context of each token in relation to the whole input. 
### Step 3: Decode autoregressively
The **decoder** generates the output **autoregressively**, which means it generates one token at a time, using the previously generated tokens as context.

## 3. Advantages & Limitations

| | Aspect | Commentary |
|--|--------|------------|
| 🟢 | **Dual architecture** | BART is a transformer encoder-decoder model with a bidirectional (BERT-like) encoder and an autoregressive (GPT-like) decoder |
| 🟢 | **Flexible noising** | A key advantage of this setup is the noising flexibility. Arbitrary transformations can be applied to the original text, including changing its length. |
| 🟢 | **State-of-the-art results** | BART achieves new state-of-the-art results on a range of abstractive dialogue, question answering, and summarization tasks, with sometime gains of up to 6 ROUGE. |
| 🟢 | **Strong text infilling** | Text infilling forces the model to estimate the appropriate length of missing spans, a task that extends beyond simple word prediction. |
| 🔴 | **Expensive reconstruction** | The full-sequence reconstruction objective is computationally expensive. Unlike T5, which generates only corrupted spans, BART's decoder processes every token of the original sequence. For long documents, this increases training cost significantly. |
| 🔴 | **No parallelized decoding** | The autoregressive decoder also prevents parallelization during training, unlike the encoder which processes all positions simultaneously.|
| 🔴 | **Fixed corruption scheme** | The fixed corruption scheme may not be optimal for all downstream tasks. BART's text infilling with sentence permutation works well for summarization and generation, but other corruptions might better suit specific applications. |
| 🔴 | **Primarily English-focused** | The base BART model is pre-trained on English only. Its multilingual variant mBART was required to extend coverage to other languages. |

**Hosted Open-weight LLMs** (via Hugging Face Inference Endpoints, Together AI, etc.) give you open-weight models like Llama or Mistral run on someone else's GPUs. 
**Self-hosted LLMs** (Llama running on your own servers) give you full control, privacy, and customization but require infrastructure and ML-ops expertise.</span>

## 1. Overview

<!-- What is this concept? 2–4 sentences that a newcomer could understand.
     Include: what problem it solves, why it exists, where it fits in the AI landscape. -->

**Proprietary LLM (ChatGPT, Claude)**: Closed-weight models accessible only through a paid API. The provider owns the weights, infrastructure, and updates. You send a prompt, get a response, and never see "inside" the model. Known for state-of-the-art reasoning quality and ease of integration — commercial models tend to lead on complex multi-step reasoning benchmarks compared to open-weight alternatives. 

**Hosted Open-weight LLM (Hugging Face)**: Open-weight models (Llama, Mistral, Qwen, etc.) whose weights anyone can download, but instead of running them yourself, a provider hosts the inference servers for you and exposes an API. This shifts you to fixed or semi-fixed GPU costs, cheaper at scale with stable workloads, while letting developers fine-tune on proprietary data and adjust model behavior more deeply. Providers like Together AI, Hugging Face Inference Endpoints, Fireworks AI, Baseten, and Groq are among the leading options in 2026. 

**Self-hosted LLM (Llama)**: You download the open-weight model and run it entirely on your own hardware (on-prem servers, your own cloud VMs, or even a laptop with tools like Ollama or llama.cpp). Your data stays within your own infrastructure and network, but you become responsible for implementing your own security controls. A key benefit is full flexibility to apply inference optimizations — frameworks like vLLM and SGLang support continuous batching and speculative decoding for your specific use case.

---
## 2. Best Use Cases

 Type | Best For |
|--------|------------|
| **Proprietary** | Fast prototyping, general-purpose assistants, complex reasoning/coding tasks, teams without ML-ops expertise |
| **Hosted Open-weight** | Custom fine-tuned models without managing GPUs; real-time apps needing consistent low-latency throughput.|
| **Self-hosted** | Processing sensitive, regulated, or proprietary data in industries like healthcare, finance, defense, or IP-heavy sectors; deep customization; long-term cost control at scale |

---

## 3. Differences

 Criterion | Proprietary| Hosted Open-weight| Self-hosted| 
|--------|------------|----------|--------|
| **Weight access** |None (black box)|Open weights, managed infra|Open weights, your infra|
| **Data privacy** |Sent to provider's servers|Sent to third-party provider|Stays in-house|
|**Cost model**|Pay-per-token|Pay-per-token or per-GPU-hour|Fixed/CapEx GPU cost|
|**Customization**|Limited (prompting, some fine-tuning APIs)|Fine-tuning supported|Full control (architecture, weights, inference stack)|**Ops effort**|None|Low|High (you manage scaling, latency, hardware)|

## 4. Code Example

> **Goal:** Show examples of how to use the different models.

```python
# ---------------------------------------------------------------------------
# Proprietary (Gemini API)
# ---------------------------------------------------------------------------

from google import genai

client = genai.Client(api_key="YOUR_GEMINI_API_KEY")

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explain LLMs simply."
)
print(response.text)

# ---------------------------------------------------------------------------
# Hosted Open-weight (Hugging Face Inference Providers)
# ---------------------------------------------------------------------------

from huggingface_hub import InferenceClient

client = InferenceClient(api_key="YOUR_HF_TOKEN")

response = client.chat.completions.create(
    model="meta-llama/Llama-3.1-8B-Instruct",
    messages=[{"role": "user", "content": "Explain LLMs simply."}],
    max_tokens=200,
)
print(response.choices[0].message.content)

# ---------------------------------------------------------------------------
# Self-hosted (Llama via Ollama)
# ---------------------------------------------------------------------------

from ollama import chat

response = chat(
    model="llama3.2",
    messages=[{"role": "user", "content": "Explain LLMs simply."}]
)
print(response.message.content)
```

## 5. Key Takeaways
<div style="font-size: 16px; line-height: 1.6;">

- **Proprietary means convenience, not ownership**
- **Open-weight ≠ free to run.** "open-weight" describes the weights, not the hosting; you still need GPUs somewhere.
- **Privacy drives the self-hosting decision**
- **Hosted open-weight is the middle ground**. Open-weight flexibility without managing your own servers.
